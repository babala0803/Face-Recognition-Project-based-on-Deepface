```python
import os
import numpy as np
from util import get_cls,ground_true_list,get_predict,show_results,get_y_sim,get_all_path
import matplotlib.pyplot as plt
from deepface import DeepFace
from sklearn.metrics import RocCurveDisplay,roc_curve
def get_y_sim(model='VGG-Face',distance_metric='cosine'):
    db_path = "./testset2/"  # 数据库路径
    relative_path = get_all_path()
    real_name_list = ground_true_list()
    y_true = []
    y_score = []
    for index, path in enumerate(relative_path):
        real_name = real_name_list[index]

        result = DeepFace.find_all(img_path=path, db_path=db_path, enforce_detection=False,model_name=model,distance_metric=distance_metric)
        for row in result[0].itertuples(index=False):
            score = row[5]
            file_name = row[0].split("/")[-1].split("_")[0]
            if real_name == file_name:
                y_true.append(1)
            else:
                y_true.append(0)
            y_score.append(1-score)
    return y_true,y_score
y_true,y_sim=get_y_sim()
y_true1,y_sim1=get_y_sim(model='Facenet')
y_true2,y_sim2=get_y_sim(model='ArcFace')
y_true3,y_sim3=get_y_sim(distance_metric='euclidean')

RocCurveDisplay.from_predictions(y_true,y_sim, name='vgg Verification')
RocCurveDisplay.from_predictions(y_true1,y_sim1, name='Facenet Verification')
RocCurveDisplay.from_predictions(y_true2,y_sim2, name='ArcFace Verification')
RocCurveDisplay.from_predictions(y_true3,y_sim3, name='euclidean Verification')
```
