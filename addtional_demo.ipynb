```python
def get_file_paths(directory):
    file_paths = []  # 存储文件路径的列表

    # 获取文件夹内的所有文件和文件夹
    items = os.listdir(directory)

    # 遍历每个项目
    for item in items:
        item_path = os.path.join(directory, item)  # 构建项目的完整路径

        # 检查是否为文件
        if os.path.isfile(item_path):
            file_paths.append(item_path)  # 将文件路径添加到列表中

    return file_paths

folder_path = './add_testset1/'  # 替换为实际的文件夹路径
file_paths = get_file_paths(folder_path)
```

```python
m = 1

df_name = pd.read_csv('./identity.csv', encoding='latin-1')

# 遍历每个项目
for original_img_path in file_paths:
    model_name = 'Facenet'
    distance_metric = 'cosine'
    df = DeepFace.find_all(img_path=original_img_path, db_path='./add_testset2/', model_name=model_name, distance_metric=distance_metric, enforce_detection=False)
    threshold = dst.findThreshold(model_name, distance_metric)
    result_df = df[0][df[0][f"{model_name}_{distance_metric}"] <= threshold]

    paint_len = np.min([result_df.shape[0], 4])

    result_df = result_df[:paint_len]

    # 创建子图
    fig, axs = plt.subplots(1, len(result_df)+1, figsize=(12, 4))

    # 绘制原始图片
    img = Image.open(original_img_path)
    axs[0].imshow(img)

    # 绘制存储地址对应的图片
    for i, identity in enumerate(result_df['identity']):
        img = Image.open(identity)
        axs[i+1].imshow(img)
        axs[i+1].set_title('Image {}'.format(i+1))

        if i==0:
            predicted_class_id = os.path.splitext(os.path.basename(identity))[0].split('_')[0]
            predicted_name = df_name[df_name['Class_ID'] == predicted_class_id][' Name'].iloc[0]

    axs[0].set_title('Predict: {}'.format(predicted_name))
            

    # 设置图像标题和间距
    fig.suptitle('Comparison of Images', fontsize=16)
    plt.tight_layout()

    # 显示图像
    plt.savefig('add_fig/result_'+str(m)+'_'+str(1)+'.png')
    # plt.show()
    plt.close()

    


    # 计算行数和列数
    num_rows = len(result_df)
    num_cols = 2

    # 创建网格规格
    spec = gridspec.GridSpec(ncols=num_cols, nrows=num_rows, width_ratios=[1, 1])

    # 创建图形对象
    fig = plt.figure(figsize=(6, 4))

    # 绘制原始图片
    img = Image.open(original_img_path)
    ax = fig.add_subplot(spec[:, 0])
    ax.imshow(img, aspect='auto')
    ax.set_title('Predict: {}'.format(predicted_name))

    embeddings = DeepFace.represent(img_path = original_img_path, model_name = model_name, enforce_detection=False)
    # draw a rectangle around the face
    face_coord = embeddings[0]['facial_area']
    rect = patches.Rectangle((face_coord['x'], face_coord['y']), 
                            face_coord['w'], face_coord['h'], 
                            linewidth = 2, 
                            edgecolor = 'r', 
                            facecolor = 'none')

    # add the patch to the axes
    ax.add_patch(rect)

    # 隐藏坐标轴
    ax.axis('off')

    # 绘制存储地址对应的图片
    for i, identity in enumerate(result_df['identity']):
        # img = Image.open(identity)
        img = DeepFace.detectFace(identity)
        ax = fig.add_subplot(spec[i, 1])
        ax.imshow(img)
    
        ax.set_title('Image {} \n dis= {:.3f}'.format(i+1, result_df.iloc[i,-1]), fontsize=8, loc='right')

        # 隐藏坐标轴
        ax.axis('off')

    # 设置图像标题和间距
    fig.suptitle('Comparison of Images', fontsize=14)
    plt.tight_layout()

    # 显示图像
    plt.savefig('add_fig/result_'+str(m)+'_'+str(2)+'.png')
    # plt.show()
    plt.close()

    m += 1
```
